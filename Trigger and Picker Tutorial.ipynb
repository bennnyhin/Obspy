```python
from obspy.core import read
from obspy.signal.trigger import plot_trigger
trace = read("https://examples.obspy.org/ev0_6.a01.gse2")[0]
df = trace.stats.sampling_rate
print(df)
```

```python
from obspy.signal.trigger import classic_sta_lta
cft = classic_sta_lta(trace.data, int(5*df), int(10*df)) #how is the length of short time average window in sample known before?
plot_trigger(trace, cft, 1.5, 0.5)
```
